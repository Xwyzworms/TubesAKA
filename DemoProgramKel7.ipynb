```
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import time
from matplotlib.pyplot import draw,show
from IPython.display import clear_output
class TubesAka:
    def __init__(self):
        print("Initialize Aka")
        self.hasilRunningBucket = []
        self.hasilRunningBubble = []
        self.udahExecute=True
        self.hasilRunningTimeBucket = {}
        self.hasilRunningTimeBubble = {}

    def getHasilRunningBucket(self):
        """Return Dictionary Hasil Running Bucket sort Berdasarkan inputan"""
        return self.hasilRunningBucket

    def getHasilRunningBubble(self):
        '''Return Dictionary Hasil Running Bubble sort Berdasarkan inputan'''
        return self.hasilRunningBubble

    def setBucket(self,hasilrunning):
        ''' Set isi hasilRunning bucket '''
        self.hasilRunningBucket=hasilrunning

    def setBubble(self, hasilrunning):
        '''Set isi hasilRunningBubble '''
        self.hasilRunningBubble = hasilrunning

    def getHasilRunningTimeBucket(self):
        """Return Dictionary Hasil Running Bucket sort Berdasarkan inputan"""
        return self.hasilRunningTimeBucket

    def getHasilRunningTimeBubble(self):
        '''Return Dictionary Hasil Running Bubble sort Berdasarkan inputan'''
        return self.hasilRunningTimeBubble

    def setTimeBucket(self,pilihan,hasilrunning):
        ''' Set isi hasilRunningTimebucket '''
        self.hasilRunningTimeBucket[pilihan]=hasilrunning

    def setTimeBubble(self,pilihan,hasilrunning):
        '''Set isi hasilRunningTimeBubble '''
        self.hasilRunningTimeBubble[pilihan] = hasilrunning

    
    def menu(self):
        """
        Menu untuk memberikan pilihan kepada user
        """
        try:
            print("Selamat datang pada menu perbandingan Kompleksitas Algoritma Bucket Sort dengan Bubble sort !!")
            print("1.EksekusiSekarang")
            print("2.Visualisasi")
            pilihan = int(input("Silahkan Masukan Inputan : "))
            clear_output()
            if (pilihan == 1):
                for i in [10,100,1000,10000,100000]:
                    if (i == 100000):
                        print("Akan Sangat lama jika ditunggu, maka dari itu digunakan waktu yang telah didapat")
                        self.setTimeBucket(100000, 0.2713597774505615)
                        self.setTimeBubble(100000, 2410.018307209015)
                        self.GenerateVisualisasi()
                    else:                             
                        np.random.seed(np.random.randint(1,10,1))
                        array = self.generateArray(i)
                        print(f"Saat ini Kamu berada pada looping dengan inputan : {i}")
                        print(f"Array yang didapat adalah \n{array[:10]}\n\n")
                        print("sedang Sorting dengan Bubble, Bentar Ya ....")
                        begin = time.time()
                        self.setBubble(self.bubblesort(array))
                        end = time.time() 
                        print("| Hasil Bubble Sort : {} |\n\n".format(self.getHasilRunningBubble()[:10],""))
                        self.setTimeBubble(i, end - begin)
                        print("sedang Sorting dengan Bucket, Bentar Ya ....")
                        begin = time.time()
                        self.setBucket(self.bucket_sort(array))
                        end = time.time()
                        print("| Hasil Bucket Sort : {} |\n\n".format(self.getHasilRunningBucket()[:10],""))
                        self.setTimeBucket(i, end-begin)
                        print(f"\nHasil Eksekusi Bucket {self.getHasilRunningTimeBucket()}")
                        print(f"\nHasil Eksekusi Bubble {self.getHasilRunningTimeBubble()}")
                        _ = self.GenerateVisualisasi()
                      
            elif (pilihan == 2):
                _=self.GenerateVisualisasi()
            self.menu()
        except ValueError as e:
            print("Sepertinya Kamu salah memasukan inputan")
            
    
    def insertion_Sort(self,arr):
        for i in range(1, len(arr)):
            key = arr[i]
            before = i - 1
            while before >= 0 and key < arr[before]:
                arr[before + 1] = arr[before]
                before = before - 1
                arr[before + 1] = key
        return arr


    def bucket_sort(self,arr,Slotbucket=10):
        '''
        I.S : Array telah Terbentuk dengan n elements dan belum ter-urut
        F.S : Array telah terurut dengan algoritma BucketSort

        Arr : Adalah Array dengan N elements
        '''

        ''' Inisiasi Jumlah bucket 
            Inisiasi Variable LokasiBucket
            Inisiasi List untuk Bucket yang telah tersorting
        '''

        buckets = [] 
        bucketLocation = 0
        Sortedbucket = []
        # Buat Bucket dengan jumlah Slotbucket
        for i in range(0,Slotbucket): 
            buckets.append([])

        # #Masukan Element yang ada pada array ke bucket yang sesuai
        for i in range(0, len(arr)): 
            bucketLocation = int( arr[i]  * Slotbucket)
            buckets[bucketLocation].append(arr[i])

        # Lakukan Sorting pada setiap Bucket
        for i in range(0, len(buckets)):
            buckets[i] = self.insertion_Sort(buckets[i])
        #Concatenate Bucket menjadi satu 
        for i in range(Slotbucket):
            for j in range(len(buckets[i])):
                Sortedbucket.append(buckets[i][j])
        
        return np.array(Sortedbucket)
    def bubblesort(self,arr):
        '''
        I.S : Array telah terbentuk dengan n elements dan belum terurut
        F.S : Array Telah Terurut dengan algoritma bubblesort

        arr : Array dengan N Elements
        '''

        for i in range(0,len(arr)):
            for j in range(0,len(arr)-i-1): 
                if arr[j] > arr[j+1]: 
                    arr[j] , arr[j+1] = arr[j+1] , arr[j] 
        return arr    
    
    def generateArray(self,Size):
        """
        Return array dengan range 0 - 1 sesuai dengan Size dari parameters
        Size: ukuran parameter
        """
        return np.random.random(Size)
    
    def GenerateVisualisasi(self):
        """
        Menampilkan Hasil visualisasi Running time Bucket sort dan bubble sort
        """
        print("Menggunakan Waktu yang telah didapat,dikarenakan running time akan sangat lama")

      
        f,ax = plt.subplots(1,1,figsize=(7,7))
        inputan = [inputs for inputs in list(self.getHasilRunningTimeBubble().keys()) ]        
        ax.plot(inputan , list(self.getHasilRunningTimeBucket().values()),
                      inputan , list(self.getHasilRunningTimeBubble().values()))
        ax.set(title="Hasil Running Time Kedua Algoritma",xlabel="Input Size",ylabel="Running time in sekon")
        ax.legend(["Bucket","Bubble"])
        show(block=False)

if __name__=="__main__":
    Yow = TubesAka()
    Yow.menu()
```
